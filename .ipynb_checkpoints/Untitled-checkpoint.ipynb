### Model Architecture and Training Strategy

#### 1. Solution Design Approach

The overall strategy for deriving a model architecture was to to study all the model paper.

I selected Nvidia model https://devblogs.nvidia.com/deep-learning-self-driving-cars/ because it was very close to use case. I haven't tried other model due to time constrain to this term, but will like to try resnet other in future.

In order to gauge how well the model was working, I split my image and steering angle data into a training and validation set(0.8 and 0.2). 

* In first attempt, My car was good in driving straight road. But, on few turns it got confuse and went on grass.
* To fix it, I added dropout layer, which worked very well on track 1 (Lake)
* When I tried on Track2 (Mountain), I observe that model was not able to sharp turns, intial four lane road with divider etc.
* I improved by adding more and more data from track2 specially for sharp turn. It improved but still could not complete the track2 successful.
* Then I observe that my data was very skewed to zero steering angle.
* I thought of removing samples with zero angle. Intially, I plan to remove sample at time of load csv file. But later I realize that model improved but it fail to work on straight road at some locations like bridge etc. To fix issue, I randomly remove different samples with steering angle=0 for each epoch. And, bingo!!. It makes my car drive on track 2 fully without stopping or getting accident.( Still at one place it goes little off road and return back on road. but I believe it can also fix with more data, My dataset was very limited with only ten of thousand samples including track1 and track2 data)
